```python
import pandas as pd
import numpy as np
from scipy import stats

import matplotlib.pyplot as plt

# Part a: Read the data and process it
file_path = "./../data//interim//klima_hamburg_TMK_TXK_TNK.csv"
data = pd.read_csv(file_path, delimiter=";")

data["DD"] = data["DD"].astype(int)
data["MM"] = data["MM"].astype(int)
data["YYYY"] = data["YYYY"].astype(int)

data["DATE"] = pd.to_datetime(
    {"year": data["YYYY"], "month": data["MM"], "day": data["DD"]}
)
data.set_index("DATE", inplace=True)
```

```python
# Filter data for months January to March (1, 2, 3)
jan_to_mar_data = data[data['MM'].isin([1, 2, 3])]
jan_to_mar_first_20 = jan_to_mar_data[jan_to_mar_data['YYYY']<= 1956]
jan_to_mar_last_20 = jan_to_mar_data[jan_to_mar_data['YYYY']>= 2003]

# Filter data for months July to September (7, 8, 9)
jul_to_sep_data = data[data['MM'].isin([7, 8, 9])]
jul_to_sep_first_20 = jul_to_sep_data[jul_to_sep_data['YYYY']<= 1956]
jul_to_sep_last_20 = jul_to_sep_data[jul_to_sep_data['YYYY']>= 2003]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# January to March mean temperatures
# Determine common bin range for January to March data
min_temp_winter_mean = min(jan_to_mar_first_20['TMK'].min(), jan_to_mar_last_20['TMK'].min())
max_temp_winter_mean = max(jan_to_mar_first_20['TMK'].max(), jan_to_mar_last_20['TMK'].max())
bins_winter_mean = np.arange(np.floor(min_temp_winter_mean), np.ceil(max_temp_winter_mean) + 0.5, 0.5)  # 0.5 degree bins

axes[0].hist(jan_to_mar_first_20['TMK'], bins=bins_winter_mean, alpha=0.5, label='1956 and earlier', color='blue')
axes[0].hist(jan_to_mar_last_20['TMK'], bins=bins_winter_mean, alpha=0.5, label='2003 and later', color='red')
axes[0].set_xlabel('Temperature (C)')
axes[0].set_ylabel('Frequency')
axes[0].set_title('January to March Mean Temperatures')
axes[0].grid(linestyle='--', alpha=0.5)
axes[0].legend()

# July to September mean temperatures
# Determine common bin range for January to March data
min_temp_summer_mean = min(jul_to_sep_first_20['TMK'].min(), jul_to_sep_last_20['TMK'].min())
max_temp_summer_mean = max(jul_to_sep_first_20['TMK'].max(), jul_to_sep_last_20['TMK'].max())
bins_summer_mean = np.arange(np.floor(min_temp_summer_mean), np.ceil(max_temp_summer_mean) + 0.5, 0.5)  # 0.5 degree bins

axes[1].hist(jul_to_sep_first_20['TMK'], bins=bins_summer_mean, alpha=0.5, label='1956 and earlier', color='blue')
axes[1].hist(jul_to_sep_last_20['TMK'], bins=bins_summer_mean, alpha=0.5, label='2003 and later', color='red')
axes[1].set_xlabel('Temperature (C)')
axes[1].set_ylabel('Frequency')
axes[1].set_title('July to September Mean Temperatures')
axes[1].grid(linestyle='--', alpha=0.5)
axes[1].legend()

plt.tight_layout()
plt.suptitle('Comparison of Mean Temperatures: Past vs Present', y=1.05, fontsize=16)
plt.show()

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# January to March mean temperatures
# Determine common bin range for January to March data
min_temp_winter_max = min(jan_to_mar_first_20['TXK'].min(), jan_to_mar_last_20['TXK'].min())
max_temp_winter_max = max(jan_to_mar_first_20['TXK'].max(), jan_to_mar_last_20['TXK'].max())
bins_winter_max = np.arange(np.floor(min_temp_winter_max), np.ceil(max_temp_winter_max) + 0.5, 0.5)  # 0.5 degree bins

axes[0].hist(jan_to_mar_first_20['TXK'], bins=bins_winter_max, alpha=0.5, label='1956 and earlier', color='blue')
axes[0].hist(jan_to_mar_last_20['TXK'], bins=bins_winter_max, alpha=0.5, label='2003 and later', color='red')
axes[0].set_xlabel('Temperature (C)')
axes[0].set_ylabel('Frequency')
axes[0].set_title('January to March Maximum Temperatures')
axes[0].grid(linestyle='--', alpha=0.5)
axes[0].legend()

# July to September mean temperatures
# Determine common bin range for January to March data
min_temp_summer_max = min(jul_to_sep_first_20['TXK'].min(), jul_to_sep_first_20['TXK'].min())
max_temp_summer_max = max(jul_to_sep_first_20['TXK'].max(), jul_to_sep_last_20['TXK'].max())
bins_summer_max = np.arange(np.floor(min_temp_summer_max), np.ceil(max_temp_summer_max) + 0.5, 0.5)  # 0.5 degree bins

axes[1].hist(jul_to_sep_first_20['TXK'], bins=bins_summer_max, alpha=0.5, label='1956 and earlier', color='blue')
axes[1].hist(jul_to_sep_last_20['TXK'], bins=bins_summer_max, alpha=0.5, label='2003 and later', color='red')
axes[1].set_xlabel('Temperature (C)')
axes[1].set_ylabel('Frequency')
axes[1].set_title('July to September Maximum Temperatures')
axes[1].grid(linestyle='--', alpha=0.5)
axes[1].legend()

plt.tight_layout()
plt.suptitle('Comparison of Maximum Temperatures: Past vs Present', y=1.05, fontsize=16)
plt.show()

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# January to March mean temperatures
# Determine common bin range for January to March data
min_temp_winter_min = min(jan_to_mar_first_20['TNK'].min(), jan_to_mar_last_20['TNK'].min())
max_temp_winter_min = max(jan_to_mar_first_20['TNK'].max(), jan_to_mar_last_20['TNK'].max())
bins_winter_min = np.arange(np.floor(min_temp_winter_min), np.ceil(max_temp_winter_min) + 0.5, 0.5)  # 0.5 degree bins

axes[0].hist(jan_to_mar_first_20['TNK'], bins=bins_winter_min, alpha=0.5, label='1956 and earlier', color='blue')
axes[0].hist(jan_to_mar_last_20['TNK'], bins=bins_winter_min, alpha=0.5, label='2003 and later', color='red')
axes[0].set_xlabel('Temperature (C)')
axes[0].set_ylabel('Frequency')
axes[0].set_title('January to March Minimum Temperatures')
axes[0].grid(linestyle='--', alpha=0.5)
axes[0].legend()

# July to September mean temperatures
# Determine common bin range for January to March data
min_temp_summer_min = min(jul_to_sep_first_20['TNK'].min(), jul_to_sep_first_20['TNK'].min())
max_temp_summer_min = max(jul_to_sep_first_20['TNK'].max(), jul_to_sep_last_20['TNK'].max())
bins_summer_min = np.arange(np.floor(min_temp_summer_min), np.ceil(max_temp_summer_min) + 0.5, 0.5)  # 0.5 degree bins

axes[1].hist(jul_to_sep_first_20['TNK'], bins=bins_summer_min, alpha=0.5, label='1956 and earlier', color='blue')
axes[1].hist(jul_to_sep_last_20['TNK'], bins=bins_summer_min, alpha=0.5, label='2003 and later', color='red')
axes[1].set_xlabel('Temperature (C)')
axes[1].set_ylabel('Frequency')
axes[1].set_title('July to September Minimum Temperatures')
axes[1].grid(linestyle='--', alpha=0.5)
axes[1].legend()

plt.tight_layout()
plt.suptitle('Comparison of Minimum Temperatures: Past vs Present', y=1.05, fontsize=16)
plt.show()
```

```python
# Calculate the statistical moments (mean, variance, skewness) for each temperature type and time period

# Create a dictionary to store the results
stats_results = {
    'Winter (Jan-Mar)': {},
    'Summer (Jul-Sep)': {}
}

# Calculate moments for winter periods (Jan-Mar)
for temp_type in ['TMK', 'TXK', 'TNK']:
    temp_name = {'TMK': 'Mean Temp', 'TXK': 'Max Temp', 'TNK': 'Min Temp'}[temp_type]
    
    # First 20 years
    first_mean = jan_to_mar_first_20[temp_type].mean()
    first_var = jan_to_mar_first_20[temp_type].var()
    first_skew = stats.skew(jan_to_mar_first_20[temp_type])
    
    # Last 20 years
    last_mean = jan_to_mar_last_20[temp_type].mean()
    last_var = jan_to_mar_last_20[temp_type].var()
    last_skew = stats.skew(jan_to_mar_last_20[temp_type])
    
    stats_results['Winter (Jan-Mar)'][temp_name] = {
        '1956 and earlier': {'Mean': first_mean, 'Variance': first_var, 'Skewness': first_skew},
        '2003 and later': {'Mean': last_mean, 'Variance': last_var, 'Skewness': last_skew},
        'Difference': {'Mean': last_mean - first_mean, 'Variance': last_var - first_var, 
                      'Skewness': last_skew - first_skew}
    }

# Calculate moments for summer periods (Jul-Sep)
for temp_type in ['TMK', 'TXK', 'TNK']:
    temp_name = {'TMK': 'Mean Temp', 'TXK': 'Max Temp', 'TNK': 'Min Temp'}[temp_type]
    
    # First 20 years
    first_mean = jul_to_sep_first_20[temp_type].mean()
    first_var = jul_to_sep_first_20[temp_type].var()
    first_skew = stats.skew(jul_to_sep_first_20[temp_type])
    
    # Last 20 years
    last_mean = jul_to_sep_last_20[temp_type].mean()
    last_var = jul_to_sep_last_20[temp_type].var()
    last_skew = stats.skew(jul_to_sep_last_20[temp_type])
    
    stats_results['Summer (Jul-Sep)'][temp_name] = {
        '1956 and earlier': {'Mean': first_mean, 'Variance': first_var, 'Skewness': first_skew},
        '2003 and later': {'Mean': last_mean, 'Variance': last_var, 'Skewness': last_skew},
        'Difference': {'Mean': last_mean - first_mean, 'Variance': last_var - first_var, 
                      'Skewness': last_skew - first_skew}
    }

# Create DataFrame for better visualization
seasons = []
temp_types = []
periods = []
moments = []
values = []

for season, season_data in stats_results.items():
    for temp_type, period_data in season_data.items():
        for period, moment_data in period_data.items():
            for moment, value in moment_data.items():
                seasons.append(season)
                temp_types.append(temp_type)
                periods.append(period)
                moments.append(moment)
                values.append(value)

stats_df = pd.DataFrame({
    'Season': seasons,
    'Temperature Type': temp_types,
    'Period': periods,
    'Moment': moments,
    'Value': values
})

# Display the results in a structured format
pd.set_option('display.precision', 3)
print("Statistical Moments of Temperature Distributions:")
for season in ['Winter (Jan-Mar)', 'Summer (Jul-Sep)']:
    print(f"\n{season}")
    for temp_type in ['Mean Temp', 'Max Temp', 'Min Temp']:
        print(f"\n{temp_type}")
        df_subset = stats_df[(stats_df['Season'] == season) & 
                              (stats_df['Temperature Type'] == temp_type)]
        # Reshape for easier comparison
        pivot_df = df_subset.pivot_table(
            index='Moment', 
            columns='Period',
            values='Value'
        )
        print(pivot_df)
```

The histograms show the distribution of daily mean, maximum, and minimum temperatures for the winter and summer seasons, comparing the first 20 years of records to the most recent 20 years. In almost all cases, a clear shift in the mean can be observed, with temperatures generally higher in the last 20 years. This shift is more pronounced during the winter months. For summer maximum temperatures, the shift in the mean is less pronounced.

Changes in the overall shape of the distributions are difficult to discern from the histograms alone. However, statistical analysis of the moments provides further insight. In winter, the variance of mean, maximum, and minimum temperatures is lower in the last 20 years, indicating more stable temperatures compared to the earlier period. In summer, the picture is more mixed: maximum temperatures exhibit increased variance in the last 20 years, while mean and minimum temperatures show only minor changes.

Additionally, all distributions exhibit a shift toward more positive skewness, indicating that the right tails of the distributions have become longer. This suggests an increased frequency of unusually high temperature values, particularly in recent decades.

```python
# Calculate the lower and upper 5% quantiles for each distribution
percentile_results = {
    'Winter (Jan-Mar)': {},
    'Summer (Jul-Sep)': {}
}

# Calculate percentiles for winter periods (Jan-Mar)
for temp_type in ['TMK', 'TXK', 'TNK']:
    temp_name = {'TMK': 'Mean Temp', 'TXK': 'Max Temp', 'TNK': 'Min Temp'}[temp_type]
    
    # First 20 years
    first_lower_5 = jan_to_mar_first_20[temp_type].quantile(0.05)
    first_upper_5 = jan_to_mar_first_20[temp_type].quantile(0.95)
    
    # Last 20 years
    last_lower_5 = jan_to_mar_last_20[temp_type].quantile(0.05)
    last_upper_5 = jan_to_mar_last_20[temp_type].quantile(0.95)
    
    percentile_results['Winter (Jan-Mar)'][temp_name] = {
        '1956 and earlier': {'Lower 5%': first_lower_5, 'Upper 5%': first_upper_5},
        '2003 and later': {'Lower 5%': last_lower_5, 'Upper 5%': last_upper_5},
        'Difference': {'Lower 5%': last_lower_5 - first_lower_5, 'Upper 5%': last_upper_5 - first_upper_5}
    }

# Calculate percentiles for summer periods (Jul-Sep)
for temp_type in ['TMK', 'TXK', 'TNK']:
    temp_name = {'TMK': 'Mean Temp', 'TXK': 'Max Temp', 'TNK': 'Min Temp'}[temp_type]
    
    # First 20 years
    first_lower_5 = jul_to_sep_first_20[temp_type].quantile(0.05)
    first_upper_5 = jul_to_sep_first_20[temp_type].quantile(0.95)
    
    # Last 20 years
    last_lower_5 = jul_to_sep_last_20[temp_type].quantile(0.05)
    last_upper_5 = jul_to_sep_last_20[temp_type].quantile(0.95)
    
    percentile_results['Summer (Jul-Sep)'][temp_name] = {
        '1956 and earlier': {'Lower 5%': first_lower_5, 'Upper 5%': first_upper_5},
        '2003 and later': {'Lower 5%': last_lower_5, 'Upper 5%': last_upper_5},
        'Difference': {'Lower 5%': last_lower_5 - first_lower_5, 'Upper 5%': last_upper_5 - first_upper_5}
    }

# Create DataFrame for better visualization
seasons = []
temp_types = []
periods = []
percentiles = []
values = []

for season, season_data in percentile_results.items():
    for temp_type, period_data in season_data.items():
        for period, percentile_data in period_data.items():
            for percentile, value in percentile_data.items():
                seasons.append(season)
                temp_types.append(temp_type)
                periods.append(period)
                percentiles.append(percentile)
                values.append(value)

percentile_df = pd.DataFrame({
    'Season': seasons,
    'Temperature Type': temp_types,
    'Period': periods,
    'Percentile': percentiles,
    'Value': values
})

# Display the results in a structured format
pd.set_option('display.precision', 2)
print("5% and 95% Percentiles of Temperature Distributions:")
for season in ['Winter (Jan-Mar)', 'Summer (Jul-Sep)']:
    print(f"\n{season}")
    for temp_type in ['Mean Temp', 'Max Temp', 'Min Temp']:
        print(f"\n{temp_type}")
        df_subset = percentile_df[(percentile_df['Season'] == season) & 
                                  (percentile_df['Temperature Type'] == temp_type)]
        # Reshape for easier comparison
        pivot_df = df_subset.pivot_table(
            index='Percentile', 
            columns='Period',
            values='Value'
        )
        print(pivot_df)
```

Both the lower and upper 5th percentiles shift to the right across all seasons for mean, minimum, and maximum temperatures. However, it is particularly noticeable that the lower 5th percentile during winter shifts significantly, by nearly 5C. In contrast, the upper 5th percentile shifts by only about 1C, which aligns with the reduced variance observed in the previous example. In summer, the overall shift is less pronounced, with shifts around 1C for all percentiles.

These observations indicate that extreme cold events are becoming rarer in winter, while extreme hot days are becoming more common, consistent with the earlier analysis that showed an increased likelihood of extreme warm days in winter. In summer, a similar trend is evident: extreme cold values are warming, and extreme heat values are also increasing.

Regarding the notion that a small shift in the mean can lead to significant changes in extreme values, we find that the distribution is indeed changing. However, the new distribution exhibits lower variance, and the shift in the upper 5th percentile is smaller than the shift in the mean. This suggests that we are not seeing an increase in extreme values solely based on the new distribution (when "extreme" is defined according to the updated distribution).

Nonetheless, because values that were extreme in the first 20 years remain extreme by traditional definitions, the frequency of extreme values is higher, simply due to the entire distribution shifting to the right. This shift is important in the context of its potential impact on ecosystems and society.
