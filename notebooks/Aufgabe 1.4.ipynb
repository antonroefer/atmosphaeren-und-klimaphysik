```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sea_level_rise(depth_of_warming, temperature):
    area = 3.61e14 #m^2
    beta = 4e-4 #1/K

    # Calculate the volume of water in the ocean that warms up
    V = area * depth_of_warming

    # Calculate the change in sea level due to thermal expansion
    delta_h = (beta * temperature * V) / (area)

    return delta_h
```

```python
# Create a meshgrid for depth and temperature values
depths_grid = np.linspace(0, 3500, 100)  # More points for smoother contour
temperatures_grid = np.linspace(0, 2.0, 100)  # More points for smoother contour

depth_grid, temp_grid = np.meshgrid(depths_grid, temperatures_grid)

# Calculate sea level rise for each combination
sea_level_grid = np.zeros_like(depth_grid)
for i in range(len(temperatures_grid)):
    for j in range(len(depths_grid)):
        sea_level_grid[i, j] = sea_level_rise(depth_grid[i, j], temp_grid[i, j])

# Create a contour plot
plt.figure(figsize=(10, 6))
contour = plt.contourf(depth_grid, temp_grid, sea_level_grid, levels=50, cmap='viridis')
plt.colorbar(contour, label='Sea Level Rise (m)')

# Add labels and title
plt.xlabel('Depth of Warming (m)')
plt.ylabel('Temperature Increase (K)')
plt.title('Sea Level Rise due to Thermal Expansion')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
